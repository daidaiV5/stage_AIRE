```python
df_empty = pd.DataFrame(columns=['strain', 'tissue', 'age', 'Sex','genotype','series_id','relation','molecule_ch1','instrument_model','SRA'])
for gsm_name, gsm in gse.gsms.items():
    index=gsm_name
    df_empty['series_id']=geo
    df_empty.loc[index] = 'NA'
    for key, value in gsm.metadata.items():
        if key == "characteristics_ch1" :
            df_empty['strain'][index]=value[0]
            df_empty['tissue'][index]=value[2]
            df_empty['age'][index]=value[1] # change each times
            #df_empty['Sex'][index]="Sex: Male"
            #df_empty['genotype'][index]=value[4]
        if key == "relation":
            df_empty['relation'][index]=value
            x=value[1].split('=')[1]
            df_empty['SRA'][index]=x
        if key == "molecule_ch1":
            df_empty['molecule_ch1'][index]=value
        if key == "instrument_model":
            df_empty['instrument_model'][index]=value
```
